```python
import datetime
import numpy as np
import os
import matplotlib.pyplot as plt
import pandas as pd
import glob
import seaborn as sns
import math
from pprint import pprint
from scipy import stats
from scipy.stats import ks_2samp
from statsmodels.stats.multitest import multipletests
from statannotations.Annotator import Annotator
```

```python
#                   *** GET TIME OF ANALYSIS START ***
# could turn this into a module?
toa = str(datetime.datetime.today()).split()                                   #Time of analysis
today = toa[0]
now = toa[1]
timestamp = today.replace('-','')+'-'+now.replace(':','')[:6]
timestamp
```

```python
#                *** WHAT TO ANALYZE // WHERE TO GET/STORE **
# Key identifiers
prep = 'CYS_013'

# Directories 
#   existing ones 
dirMain = '/Users/joyfranco/Dropbox (Partners HealthCare)/JF_Shared/Data/CYS/'
dirPrep = dirMain+prep+'/'
dirTIA = dirPrep+'TIA_MeasureROIs.Results/'
dirRVs = dirTIA+'ROIVals/'
dirMD = dirPrep+'Metadata/'

#   ones that need to be made    
dirRes = dirTIA+"AnalysisResults/"
```

```python
# Create directory for storing plots with identified peaks 
if not os.path.exists(dirRes): 
    os.mkdir(dirRes)
```

```python
#    *** LOAD THE SPREADSHEETS FROM MEMORY ***
dfMDI = pd.read_csv(dirMD+prep+'.Metadata.Imaging.20x.csv')
dfMDS = pd.read_csv(dirMD+prep+'.Metadata.Samples.csv')
```

```python
# *** COMPILE INFORMATION ABOUT ROIS FROM MEASUREMENT FILES ***
# Get the list of all ROI mean grey value measurement files to evaluate
os.chdir(dirRVs)
flRVs = glob.glob('*.csv')
print("These are the measurement files available for analysis")
flRVs
```

```python
# Initiatlize dataframe for storing roi info
dfRVs = pd.DataFrame()        

# Iterate through the files for each sample and generate a list to rois 
#   to exclude based on size
for file in flRVs:
    # Setup imagename
    im = file.removesuffix('.ROIVals.csv')
    # *** Get the metadata for this image ***
    # Get the index of the sample & retrieve metadata 
    inI = dfMDI.index[dfMDI['Filename']== im].tolist()[0]
    animID = dfMDI.loc[inI]['AnimalID']
    inS = dfMDS.index[dfMDS['animalID']== animID].tolist()[0]

    # Read in the ROI measurements for this image 
    df = pd.read_csv(dirRVs+file)

    # Add metadata to the ROI measurements
    df['animalID'] = dfMDI.loc[inI]['AnimalID']
    df['slides'] = dfMDS.loc[inS]['slides']
    df['column'] = dfMDS.loc[inS]['column']
    df['row'] = dfMDS.loc[inS]['row']
    df['genotype'] = dfMDS.loc[inS]['genotype']
    df['group'] = dfMDS.loc[inS]['group']
    df['sex'] = dfMDS.loc[inS]['sex']
    df['dob'] = dfMDS.loc[inS]['dob']
    df['sacDate'] = dfMDS.loc[inS]['sacDate']
    df['ageAtSac'] = dfMDS.loc[inS]['ageAtSac']
    df['collectionPrep'] = dfMDS.loc[inS]['collectionPrep']

    # Aggregate measurements across all images
    dfRVs = pd.concat([dfRVs, df], axis=0)

# Save the percent change values for the sample
dfRVs.to_csv(dirRes+prep+'.TIA_MeasureROIs.'+timestamp+'.csv')
```

```python
# *** GENERATE PLOTS ***
```

```python
plot = 'ROIMeanGrey.HABclw'
plottype = 'BoxSwarm'
title = 'Mean Grey ROI Values | HA Bclw'
x = 'group'
y = 'c1_grey'
order = ['HABclw_C1','HABclw_C2', 'HABclw_C3']
colors = {order[0]:'teal', order[1]:'turquoise', order[2]:'darkturquoise'}
hue = 'group'
markersize = 3
xlab = 'AAV HABclw Concentration'
ylab = 'ROI Mean Pixel Intensity (Au)'
ax1 = sns.swarmplot(data=dfRVs[(dfRVs['group']=='HABclw_C1') | (dfRVs['group']=='HABclw_C2') | (dfRVs['group']=='HABclw_C3')],
                    x=x,y=y, hue=hue, palette=colors, order=order, size=markersize, dodge = False)
sns.boxplot(data=dfRVs[(dfRVs['group']=='HABclw_C1') | (dfRVs['group']=='HABclw_C2') | (dfRVs['group']=='HABclw_C3')],
            x=x, y=y, hue=hue, palette=colors, order=order, boxprops={'alpha': 0.1}, ax=ax1,dodge = False,
           linewidth=.5)
ax1.get_legend().remove()
ax1.set(xlabel=xlab)
ax1.set(ylabel=ylab)
plt.title(title)

# *** SAVE THE FIGURE ***
plt.savefig(dirRes+prep+'.TIA_MeasureROIs.'+timestamp+'.'+plot+'.'+plottype+'.png',  bbox_inches='tight', dpi = 300,
            format = 'png')
```
